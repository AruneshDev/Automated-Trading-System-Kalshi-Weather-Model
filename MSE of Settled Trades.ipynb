### Austin Prediction vs Actual


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Read your CSV data into a DataFrame
df = pd.read_csv('SettledPredictions/Austin Settled Predictions.csv')

# Ensure the 'Date' column is in the datetime format and sort the DataFrame by this column
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)

# Calculate the MSE for each prediction and add it as a new column
df['MSE'] = (df['My prediction'] - df['Actual Maximum Temperature']) ** 2

# Plotting
plt.figure(figsize=(15, 5))
plt.plot(df['Date'], df['My prediction'], label='My Prediction', marker='o')
plt.plot(df['Date'], df['Actual Maximum Temperature'], label='Actual Maximum Temperature', marker='x')
plt.title('Temperature Predictions vs Actuals')
plt.xlabel('Date')
plt.ylabel('Temperature (°F)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Compute the overall MSE for the dataset
overall_mse = mean_squared_error(df['My prediction'], df['Actual Maximum Temperature'])
print(f"Overall Mean Square Error: {overall_mse}")
```
